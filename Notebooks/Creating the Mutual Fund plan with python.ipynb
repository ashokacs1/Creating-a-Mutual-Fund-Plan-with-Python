### Let’s look at the companies with the highest risks for investing:

```python
company = data.columns[1:]
votality = data[company].std()
votality.sort_values(ascending = False).head(10) 
```

```python
votality.sort_values(ascending = True).head(10) 
```

### Now, let’s look at the companies with the highest growth rate for investing:

```python
growth_company = data[company].pct_change()*100
average_growth = data[company].mean()
average_growth.sort_values(ascending=False).head(10)
```

```python
average_growth.sort_values(ascending=True).head(10)
```

### Now, let’s have a look at the companies with the highest return on investments:

```python
intial_value = data[company].iloc[0]
final_value  = data[company].iloc[-1]
return_on_investment = ((final_value-intial_value)/intial_value)*100
return_on_investment.sort_values(ascending=False).head(10)
```

### Creating a Mutual Fund Plan Based on High ROI and Low Risk

```python
return_on_investment_thresold = return_on_investment.median()
votality_thresold = votality.median()
slection_of_companies = return_on_investment[(return_on_investment>return_on_investment_thresold)&(votality<votality_thresold)]
slection_of_companies.sort_values(ascending=False)
```

### To balance the investment between these companies, we can use an inverse volatility ratio for allocation. Companies with lower volatility will get a higher weight. Let’s calculate the weight for each company:

```python
selcted_comapny = votality[slection_of_companies.index]
inverse_votality = 1/selcted_comapny
investment_ratio = (inverse_votality/inverse_votality.sum())*100
investment_ratio.sort_values(ascending=False)
```

### Analyzing Our Mutual Fund Plan

```python
top_growth_comapines = average_growth.sort_values(ascending=False).head(10)
risk_of_all_growth_companies = votality[top_growth_comapines.index]
risk_of_all_mutual_fund_comapines = votality[selcted_comapny.index]
```

```python
companies = risk_of_all_mutual_fund_comapines.index
companies_growth = risk_of_all_growth_companies.index 
mutual_fund_values =  risk_of_all_mutual_fund_comapines
growth_rate = risk_of_all_growth_companies 

plt.figure(figsize=(10,8))
plt.barh(companies,mutual_fund_values,color='blue',label='Mutual Fund Companies')
plt.barh(companies_growth,growth_rate,color='green',alpha=0.7,label='Growth Fund Comapnies')
plt.title('Risk Comaprsion: Mutual Fund vs Growth Fund Comapnies')
plt.xlabel('Votality in % ')
plt.ylabel('Comapnies')
plt.legend(title='Type of Company')
plt.show()
```

### Now, let’s compare the ROI of both the groups as well:

```python
roi_of_all_growth_companies = return_on_investment[top_growth_comapines.index]
roi_of_all_mutual_fund_companies = return_on_investment[selcted_comapny.index]


companies = roi_of_all_mutual_fund_companies.index
companies_growth = roi_of_all_growth_companies.index


mutual_fund_roi = roi_of_all_mutual_fund_companies
growth_rate_roi = roi_of_all_growth_companies

plt.figure(figsize=(10, 8))
plt.barh(companies, mutual_fund_roi, color='blue', label='Mutual Fund Companies')
plt.barh(companies_growth, growth_rate_roi, color='green', alpha=0.7, label='Growth Fund Companies')


plt.title('ROI Comparison: Mutual Fund vs Growth Fund Companies')
plt.xlabel('Return on Investment (%)')
plt.ylabel('Companies')

plt.legend()
plt.show()

```

### The comparison between the risk (volatility) and expected ROI for mutual fund companies (in blue) and growth rate companies (in green) shows a clear trade-off. Mutual fund companies offer lower volatility, meaning they are less risky, but also provide lower expected returns. In contrast, growth rate companies demonstrate higher volatility, indicating more risk, but they offer much higher potential returns, especially companies like Bajaj Auto and Bajaj Finserv. This highlights a common investment dilemma: lower risk comes with a lower reward, while higher risk could yield higher returns.

### For long-term investments, the goal is typically to find companies that offer a balance of stable returns and manageable risk. The companies in our mutual fund exhibit low volatility, meaning they are less risky, and their moderate returns make them solid choices for long-term, stable growth. They are well-suited for conservative investors who want steady returns without significant fluctuations in value.

# Calculating Expected Returns


```python
import numpy as np

monthly_investment = 5000 # Monthly investment in INR
years = [1, 3, 5, 10]  # Investment periods (in years)
n = 12  # Number of times interest is compounded per year (monthly)

avg_roi = roi_of_all_mutual_fund_companies.mean() / 100 

def future_value(P, r, n, t):
    return P * (((1 + r/n)**(n*t) - 1) / (r/n)) * (1 + r/n)

future_values = [future_value(monthly_investment, avg_roi, n, t) for t in years]
```
